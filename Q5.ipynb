```python
# Install necessary libraries if not already installed
# !pip install yfinance matplotlib

import yfinance as yf
import matplotlib.pyplot as plt

# Step 1: Download Tesla (TSLA) stock data
def fetch_tesla_data():
    tesla_data = yf.download('TSLA', start='2020-01-01', end='2024-01-01')
    return tesla_data

# Step 2: Define the make_graph function to plot stock data
def make_graph(stock_data, title='Tesla Stock Price'):
    plt.figure(figsize=(10, 6))
    
    # Plot the Closing prices
    plt.plot(stock_data.index, stock_data['Close'], label='Closing Price', color='b', linestyle='-', linewidth=1.5)
    
    # Add graph title and labels
    plt.title(title, fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price (USD)', fontsize=12)
    
    # Add grid and legend
    plt.grid(True)
    plt.legend()
    
    # Show the graph
    plt.show()

# Step 3: Fetch the Tesla stock data
tesla_data = fetch_tesla_data()

# Step 4: Plot the Tesla stock graph using make_graph function
make_graph(tesla_data, title='Tesla Stock Price (2020 - 2024)')
```
