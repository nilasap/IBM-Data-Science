```python
# Install necessary libraries if not already installed
# !pip install yfinance matplotlib

import yfinance as yf
import matplotlib.pyplot as plt

# Step 1: Download GameStop (GME) stock data
def fetch_gme_data():
    gme_data = yf.download('GME', start='2020-01-01', end='2024-01-01')
    return gme_data

# Step 2: Define the make_graph function to plot stock data
def make_graph(stock_data, title='GameStop Stock Price'):
    plt.figure(figsize=(10, 6))
    
    # Plot the Closing prices
    plt.plot(stock_data.index, stock_data['Close'], label='Closing Price', color='g', linestyle='-', linewidth=1.5)
    
    # Add graph title and labels
    plt.title(title, fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price (USD)', fontsize=12)
    
    # Add grid and legend
    plt.grid(True)
    plt.legend()
    
    # Show the graph
    plt.show()

# Step 3: Fetch the GameStop stock data
gme_data = fetch_gme_data()

# Step 4: Plot the GameStop stock graph using make_graph function
make_graph(gme_data, title='GameStop Stock Price (2020 - 2024)')
```
